```python
import pandas as pd
import numpy as np
exam=pd.ExcelFile('examens.xlsx')
r=exam.sheet_names
Df=exam.parse("Resultats")
Df.head()
```

#### 2. En se basant sur les tableaux des coefficients, écrire un programme qui cacule la moyenne de chaque UE et détermine si l’étudiant l’a validée ou pas.


#### EC Phy1 Phy2 TPPhy Math1 Math2 Chimie1 Chimie2 
#### Coefs 2.5 2.5 1 2.5 2.5 2.5 2.5
#### EC TPChimie Info1 Info2 Base1 Base2 Base3 Gr1 Gr2
#### Coefs 1 2.5 2.5 1.5 1.5 1 2 2


```python
Moy_Phy=(Df.Physique1*2.5 + Df.Physique2*2.5 + Df.TP_Physique)/6
Moy_Mat=(Df.Maths1*2.5 + Df.Maths2*2.5)/5
Moy_Chi=(Df.Chimie1*2.5+Df.Chimie2*2.5 + Df.TP_Chimie)/6
Moy_Inf=(Df.Info1*2.5 + Df.Info2*2.5)/5
Moy_Base=(Df.Base1*1.5 + Df.Base2*1.5 + Df.Base3)/4
Moy_Gr=(Df.Graphique1*2 + Df.Graphique2*2)/4
```

```python
Df.insert(5,'Moy_phy',Moy_Phy)
Df.insert(8,'Moy_Mat',Moy_Mat)
Df.insert(12,'Moy_Chi',Moy_Chi)
Df.insert(15,'Moy_Inf',Moy_Inf)
Df.insert(19,'Moy_Base',Moy_Base)
Df.insert(22,'Moy_Gr',Moy_Gr)
```

#### 3. Écrire un programme qui insère une colonne de validation (V : UE validée, NV : UE non validée) après chaque UE.

```python
UE_Phy=[]
for i in Moy_Phy:
    if i<10:        
        UE_Phy.append('NV')
    else:
        UE_Phy.append('V')      
UE_Mat=[]
for i in Moy_Mat:
    if i<10:        
        UE_Mat.append('NV')
    else:
        UE_Mat.append('V')      
UE_Chi=[]
for i in Moy_Chi:
    if i<10:        
        UE_Chi.append('NV')
    else:
        UE_Chi.append('V')      
UE_Inf=[]
for i in Moy_Inf:
    if i<10:       
        UE_Inf.append('NV')
    else:
        UE_Inf.append('V')
UE_Base=[]
for i in Moy_Base:
    if i<10:        
        UE_Base.append('NV')
    else:
        UE_Base.append('V') 
UE_Gr=[]
for i in Moy_Gr:
    if i<10:        
        UE_Gr.append('NV')
    else:
        UE_Gr.append('V')      
```

```python
Df.insert(6,'UE_phy',UE_Phy)
Df.insert(10,'UE_Mat',UE_Mat)
Df.insert(15,'UE_Chi',UE_Chi)
Df.insert(19,'UE_Inf',UE_Inf)
Df.insert(24,'UE_Base',UE_Base)
Df.insert(28,'UE_Gr',UE_Gr)
```

```python
T_Phy=[]
for i in UE_Phy:
    if i=='V':
        T_Phy.append(12)
    else:
        T_Phy.append(0)
T_Mat=[]
for i in UE_Mat:
    if i=='V':
        T_Mat.append(10)
    else:
        T_Mat.append(0)
T_Chi=[]
for i in UE_Chi:
    if i=='V':
        T_Chi.append(12)
    else:
        T_Chi.append(0)
T_Inf=[]
for i in UE_Inf:
    if i=='V':
        T_Inf.append(10)
    else:
        T_Inf.append(0)
T_Base=[]
for i in UE_Base:
    if i=='V':
        T_Base.append(8)
    else:
        T_Base.append(0)
T_Gr=[]
Total_Credit=[]
for i in UE_Gr:
    if i=='V':
        T_Gr.append(8)
    else:
        T_Gr.append(0)
for i in range(len(T_Base)):
    T=T_Phy[i] + T_Mat[i] + T_Chi[i] + T_Inf[i] + T_Base[i] + T_Gr[i]
    Total_Credit.append(T)
Df.insert(29,'Total_Credit',Total_Credit)
```

#### 4. Écrire un programme qui trie les noms et prénoms par ordre alphabétique.

```python
tri_Df=Df.sort_values(by=['Total_Credit','Nom','Prénom'],ascending=[False,True,True])
tri_Df
```

#### 5. Écrire un programme qui enregistre dans un nouveau fichier excel les résultats obtenus.


```python
import pandas as pd
# Cette fonction créera le fichier.
writer = pd.ExcelWriter("Cas_pratique_2.xlsx", engine="xlsxwriter")
# Ecriture du DataFrame dans le fichier : tri_df
tri_Df.to_excel(writer, "Deliberation",index=False)#CREATION DUNE FEUILLE "Deliberation".
# Enregistrement du résultat.
writer.save()
```

#### 7. Écrire un programme qui détermine pour chaque UE, le nombre d’étudiants validés. Les résultats obtenus seront rangés dans une liste.

#### 8. Tracer la courbe des UE validées en fonction du nombre d’étudiants.
